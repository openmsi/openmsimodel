2 engines are available for visualizations: "yfiles" and "cytoscape".

- "yfiles" launches yWorks' licensed graph diagram visualization widget available at [yfiles-jupyter-graphs](https://github.com/yWorks/yfiles-jupyter-graphs). It supports reading from various graph sources, restructuring graph layouts, searching by text across the entire graph, and accessing nodes and their neighbors, among other features.

- "cytoscape" launches an in-house visualization widget built with Cystoscape that is currently under development. It supports graphML sources, displays colored nodes and directed edges, and allows access to node content by clicking. Please note that this tool is unlicensed and is subject to continuous improvement.
